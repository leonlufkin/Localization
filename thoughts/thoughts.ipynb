```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 99
xi = 99
batch_size = 50000
l_star = np.random.randint(low=0, high=L, size=batch_size) #+ 1
Ts = np.random.randint(low=1, high=xi, size=batch_size)

X = np.zeros((batch_size, L))
for i, (ls, T) in enumerate(zip(l_star, Ts)):
    X[i, ls:min(ls+T,L)] = 1
    if ls+T > L:
        X[i, :ls+T-L] = 1
```

```python
a = 0
p = X[X[:, a] == 1].mean(axis=0)
_ = plt.plot(p)
_ = plt.ylim(0, 1)
_ = plt.vlines(a, 0, 1, color='r')
_ = plt.vlines((L+a+xi-1) % L, 0, 1, color='b')
_ = plt.vlines((L+a-xi+1) % L, 0, 1, color='b')
_ = plt.xlabel('Position l')
_ = plt.vlines(a + L/2, 0, 1, color='g')
```
